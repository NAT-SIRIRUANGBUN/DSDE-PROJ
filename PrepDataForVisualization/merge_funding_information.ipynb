```python
import pandas as pd
```

```python
funding_organization_interest = pd.read_csv('./funding_organization_interest.csv')
country_code = pd.read_csv('./country_code_name.csv')
```

```python
funding_organization_interest
```

```python
funding_organization_interest = funding_organization_interest.merge(country_code, how='inner')
```

```python
funding_organization_interest.reset_index().groupby(['Year', 'Funding_Agency_Country_Name', 'Subject_Field']).size().reset_index(name='count').sort_values('count', ascending=False).reset_index().drop(columns='index', axis=1)

funding_organization_interest.to_csv('funding_information_by_country_and_year.csv')
```
